```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.histplot(dataset[["CoapplicantIncome"]])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome-min"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title("Before Scaling")
sns.histplot(dataset["CoapplicantIncome"], kde=True)

plt.subplot(1, 2, 2)
plt.title("After Scaling")
sns.histplot(dataset["CoapplicantIncome-min"], kde=True)

plt.tight_layout()
plt.show()
```
